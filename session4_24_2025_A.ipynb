```python
import statsmodels.api as sm

# features = data[['Year_Type', 'Year', 'Overtime', 'Other_Salaries',
#        'Total_Salary', 'Health/Dental', 'Other_Benefits',
#        'Total_Benefits']]

features = data[['Year_Type', 'Year', 'Salaries', 'Overtime', 'Other_Salaries',
       'Total_Salary', 'Retirement', 'Health/Dental', 'Other_Benefits',
       'Total_Benefits']]

updated_features = sm.add_constant(features)
y = data[['Total_Compensation']]

model = sm.OLS(y, updated_features).fit() # Ordinary Least Squares

print(model.summary())
```

```python
data.columns
```

```python
import matplotlib.pyplot as plt
plt.scatter(model.predict(), model.resid)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
# features = ['Year_Type', 'Year', 'Overtime', 'Other_Salaries',
#        'Total_Salary', 'Health/Dental', 'Other_Benefits',
#        'Total_Benefits']

features = ['Year_Type', 'Year',  'Overtime', 
     'Health/Dental', 'Other_Benefits',
       'Total_Benefits']
for col in features:
  features = list(set(features) - set([col]))
  new_x = data[features]
  new_y = data[col]

#   model = LinearRegression()

#   model.fit(new_x, new_y)

#   pred = model.predict(new_x)

#   print(col, "vs", features, r2_score(new_y, pred))

  model_new = sm.OLS(new_y, sm.add_constant(new_x)).fit()
  print(col, ":", model_new.rsquared)
```

```python
import statsmodels.stats.api as sms
from statsmodels.compat import lzip
name = ['f-value', 'p-value']
test = sms.het_breuschpagan(model.resid, model.model.exog)
lzip(name, test[2:])
```
