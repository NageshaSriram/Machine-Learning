```python
def nearest_int(num):
  pre = int(num)
  next_n = pre + 1
  fractional_part = num - pre

  if fractional_part >= 0.5:
    return next_n
  else:
    return pre
  
```

```python
scaler = MinMaxScaler()
array = scaler.fit_transform(data_float)
data_float_scaled = pd.DataFrame(array, columns=data_float.columns)
```

```python
# SNS 
# plt.boxplot(data_float_scalled)
sns.boxplot(data_float_scaled)
```

```python
def refill(number, upper_limit, lower_limit):

  if number > upper_limit:
    return upper_limit
  elif number < lower_limit:
    return lower_limit
  else:
    return number

for col in data_float_scaled.columns:
  q1 = np.quantile(data_float_scaled[col], 0.25) # first quantile
  q3 = np.quantile(data_float_scaled[col], 0.75)

  iqr = q3 - q1

  upper_limit = q3+1.5*iqr
  lower_limit = q1-1.5*iqr

  data_float_scaled[col] = data_float_scaled[col].apply(lambda x: refill(x, upper_limit, lower_limit))
```
