```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# first data without any error
x = np.array(range(10))
y= 5 * x + np.random.normal(0,2, 10)

df = pd.DataFrame({"x": x, "y": y})
plt.scatter(x, y)
model = LinearRegression()
model.fit(df[["x"]], df[["y"]])


# outlier point
df.loc[df.index[-1]+1, :] = [0, 100]

model2 = LinearRegression()
model2.fit(df[["x"]], df[["y"]])

plt.scatter(df["x"], df["y"])
plt.plot(df["x"], model2.predict(df[["x"]]), color="red", label="linear_model2")
plt.plot(df["x"], model.predict(df[["x"]]), color="blue", label="linear_model")
plt.legend(loc="best")
plt.show()
```
